```python
#The Game of Gemstones
#Inspired by mining rocks in Shilo Village
#The game of gemstones involves populating grids randomly with gemstones
#Each gemstone is given an integer value by which it appears on the grid
#The assigned probability weights of the gemstones must equal 1
#The basic game of gemstones is 3 Gemstones (Emeralds, Rubies and Diamonds), a 4x7 Grid, and weights of [.5, .3, .2]
import math
import cmath
import decimal as dcm
import numpy as nmp
import sympy as smp
import pandas as pnd
import matplotlib as mpl

#Generate a game of Gemstones
#g is the number of gemstone options (ie 3 = emerald, diamond, ruby)
#M is the map, taken as a  list of integers ([3,2] is a 3x2 grid on which the game is played)
#W is the list of weights assigned to the gemstones. The length of this list must be equal to the number of gemstones in the game.
#S is the number of games you'd like to generate, as an integer
#Returns a PandaS Series containing the generated games 
def generate(g, sizeY, sizeX, W, s):
    #To generate a 1-D array of integers: nmp.arange([start,] stop, [step,]dtype=None, like=None)
    gems = nmp.arange(g)
    mapSize = (sizeY, sizeX)
    weights = W
    numberGames = s
    
    #List of finished games
    #List of numpy array objects, each with dimension of mapSize, populated with integers assigned to the gems
    gamesList = []

    #Generate games
    gameCounter = 0
    while gameCounter < numberGames:
        #Using nmp.random.choice to generate Y lines with length X, (Y, X) Ïµ M
        game = nmp.random.choice(gems, size=mapSize, replace=True, p=weights)
        gamesList.append(game)
        gameCounter += 1

    #Pandas Series of Games, numbered
    gamesSeries = pnd.Series(gamesList)
    
    return gamesSeries
```
